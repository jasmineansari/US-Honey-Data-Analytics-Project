## Analyzing the inferences from the data

Which states are contributing rarely for honey production since last 27 years

```python
data=df["state"].value_counts()
data.values
```

```python
plt.figure(figsize=(13,13))
plt.pie(data.values,labels=data.index,autopct="%0.2f%%")
plt.show()
```

```python
#Inference:
#Top 5 states contributing less in last 27 years are:
#  NewMexico        18
#  Nevada           15
#  SouthCarolina    12
#  Oklahoma          9
#  Maryland          9
```

### Which are the top 5 Honey producing states?

```python
plt.figure(figsize=(15,10))
sns.barplot(x=df["state"],y=df["production"],ci=None)
plt.xticks(rotation=90)
plt.show()
```

```python
new_df=df.groupby("state").sum().reset_index()
```

```python
new_df=new_df.sort_values("production",ascending=False)
```

```python
new_df
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x=new_df["state"],y=new_df["production"],ci=None)
plt.xticks(rotation=90)
plt.show()
```

### Change in mean Average price of Honey from 1995 to 2021

```python
df1=df.groupby("year").mean()['average_price'].reset_index()
```

```python
df1
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x=df1["year"],y=df1["average_price"],ci=None)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.plot(df1["year"], df1["average_price"])
plt.xticks(rotation=90)
plt.show()
```

### Which was the year when production of Honey in the whole US was the highest?

```python
df2=df.groupby("year").sum()["production"].reset_index()
```

```python
df2
```

```python
#Trend of honey productions
plt.figure(figsize=(15,10))
plt.plot(df2["year"], df2["production"])
plt.xticks(rotation=90)
plt.show()
```

```python
df2=df2.sort_values("production",ascending=False)
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x=df2["year"],y=df2["production"],ci=None)
plt.xticks(rotation=90)
plt.show()
```

### Which states are having the highest no. of colonies in year 2000?

```python
plt.figure(figsize=(15,10))
sns.barplot(x=df3["state"],y=df3["colonies_number"],ci=None)
plt.xticks(rotation=90)
plt.show()
```
